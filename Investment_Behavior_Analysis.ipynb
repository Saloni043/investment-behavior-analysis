```python
import pandas as pd

url = "https://raw.githubusercontent.com/DanieltheAnalyst1/Financial_Data_Analysis-SQL/main/Finance_data.csv"
df = pd.read_csv(url)
df.info()
```

```python
print(df.columns.tolist())
```

```python
import plotly.express as px

gender_counts = df['gender'].value_counts().reset_index()
gender_counts.columns = ['gender', 'count']

fig = px.pie(gender_counts, names='gender', values='count',
             title='Gender Distribution of Survey Participants',
             color_discrete_sequence=['#FF69B4', '#4682B4'])
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
fig2 = px.bar(gender_counts, x='gender', y='count',
              title='Gender Distribution - Bar Chart',
              color='gender', text='count')
fig2.show()
```

```python
import pandas as pd
from IPython.display import display, HTML

numerical_cols = ['age', 'Mutual_Funds', 'Equity_Market', 'Debentures',
                   'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']

desc_stats = df[numerical_cols].describe().T
desc_stats['median'] = df[numerical_cols].median()
desc_stats = desc_stats[['mean', 'median', 'std', 'min', 'max']]

styled_stats = desc_stats.style.background_gradient(cmap='Blues', subset=['mean', 'median'])\
    .format(precision=2)\
    .set_caption("Descriptive Statistics: Age & Investment Preference Ranks")\
    .set_table_styles([
        {'selector': 'th, td', 'props': [('border', '1px solid #ddd'), ('padding', '8px'), ('text-align', 'center')]},
        {'selector': 'caption', 'props': [('font-size', '16px'), ('font-weight', 'bold'), ('color', '#2F4F4F'), ('padding-bottom', '10px')]},
        {'selector': 'th', 'props': [('background-color', '#4682B4'), ('color', 'white')]}
    ])

display(styled_stats)
```

```python
avenue_counts = df['Avenue'].value_counts().reset_index()
avenue_counts.columns = ['Avenue', 'Count']

print(avenue_counts)

fig = px.bar(avenue_counts, x='Avenue', y='Count',
             title='Most Preferred Investment Avenue',
             color='Avenue', text='Count')
fig.show()
```

```python
avg_ranks = df[['Mutual_Funds', 'Equity_Market', 'Debentures',
                 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']].mean().sort_values()

print("Average Preference Rank (lower = more preferred):")
print(avg_ranks)

fig2 = px.bar(x=avg_ranks.index, y=avg_ranks.values,
              title='Average Preference Rank by Investment Type (Lower = More Preferred)',
              labels={'x': 'Investment Type', 'y': 'Average Rank'})
fig2.show()
```

```python
savings_counts = df['What are your savings objectives?'].value_counts().reset_index()
savings_counts.columns = ['Savings_Objective', 'Count']

print(savings_counts)

fig = px.bar(savings_counts, x='Savings_Objective', y='Count',
             title='Main Savings Objectives Among Participants',
             color='Savings_Objective', text='Count')
fig.show()
```

```python
source_counts = df['Source'].value_counts().reset_index()
source_counts.columns = ['Source', 'Count']

print(source_counts)

fig = px.bar(source_counts, x='Source', y='Count',
             title='Sources of Investment Information',
             color='Source', text='Count')
fig.show()
```

```python
fig2 = px.pie(source_counts, names='Source', values='Count',
              title='Information Sources - Proportion')
fig2.update_traces(textinfo='percent+label')
fig2.show()
```

```python
duration_counts = df['Duration'].value_counts().reset_index()
duration_counts.columns = ['Duration', 'Count']
print(duration_counts)

# Map duration categories to approximate numeric years (midpoint)
duration_map = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}

df['Duration_Years'] = df['Duration'].map(duration_map)

avg_duration = df['Duration_Years'].mean()
print(f"\nAverage Investment Duration: {avg_duration:.2f} years")
```

```python
fig = px.bar(duration_counts, x='Duration', y='Count',
             title='Investment Duration Distribution',
             color='Duration', text='Count')
fig.show()
```

```python
import pandas as pd

corr_cols = ['age', 'Duration_Years', 'Mutual_Funds', 'Equity_Market',
             'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']

corr_matrix = df[corr_cols].corr()

# Styled box-like table with color gradient
styled_corr = corr_matrix.style.background_gradient(cmap='RdBu_r', vmin=-1, vmax=1)\
    .format(precision=2)\
    .set_caption("Correlation Matrix: Age, Duration & Investment Preferences")\
    .set_properties(**{'border': '1px solid #ddd', 'text-align': 'center'})

styled_corr
```

```python
fig2 = px.scatter(df, x='age', y='Duration_Years',
                   title='Age vs Investment Duration',
                   trendline='ols')
fig2.show()
```

```python
from IPython.display import display, HTML

reason_cols = ['Reason_Equity', 'Reason_Mutual', 'Reason_Bonds', 'Reason_FD']

colors = ['#4682B4', '#2E8B57', '#B8860B', '#8B4789']

html = """
<div style="border:2px solid #333; border-radius:10px; padding:20px; background-color:#fafafa; width:fit-content; color:#000;">
<h3 style="margin-top:0; color:#2F4F4F;">💡 Reasons for Investment by Instrument</h3>
"""

for i, col in enumerate(reason_cols):
    counts = df[col].value_counts()
    color = colors[i % len(colors)]
    html += f"""
    <div style="margin-bottom:15px; padding:10px; border-left:5px solid {color}; background-color:#ffffff;">
    <h4 style="margin:0 0 8px 0; color:{color};">{col.replace('_', ' ')}</h4>
    <ul style="margin:0; padding-left:20px;">
    """
    for reason, count in counts.items():
        html += f"<li style='color:#000000; margin-bottom:4px;'>{reason}: <b style='color:{color};'>{count}</b></li>"
    html += "</ul></div>"

html += "</div>"
display(HTML(html))
```

```python
import plotly.express as px
import pandas as pd

reason_summary = pd.DataFrame({
    col: df[col].value_counts() for col in reason_cols
}).fillna(0)

fig = px.bar(reason_summary.T, barmode='group',
             title='Reasons for Investment by Instrument Type',
             labels={'value': 'Count', 'index': 'Investment Type'})
fig.show()
```
